Purpose: The purpose of this assignment is to train the model to detect whether or not an asteroid is hazardous

To begin, we want to import any necessary libraries, mount the goodle drive, and import the data from the file folder. Then we print out the data to see which variables could be dropped, and which are necessary in defining an asteroid as hazardous.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the dataset
import pandas as pd
data = pd.read_csv('/content/drive/My Drive/AI_ML_Datasets/nasa_asteroid.csv')

data.info()
```

Drop any variables that are not neccessary including measurements in any other unit but miles (for efficiency).

```python
data = data.drop(['Name', 'Est Dia in KM(min)', 'Est Dia in KM(max)', 'Est Dia in M(min)', 'Est Dia in M(max)', 'Est Dia in Feet(min)', 'Est Dia in Feet(max)', 'Miss Dist.(Astronomical)',
'Miss Dist.(kilometers)', 'Orbiting Body', 'Orbit Determination Date', 'Equinox', 'Close Approach Date', 'Epoch Date Close Approach'], axis=1)
```

Split the data into test, validation, and training data sets to prepare to train the model

```python
# Prepare to split the data set
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# Split the dataset into training, validation, and testing sets -- randomly shuffle the sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)

#  Split the training into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
```

Below, we are scaling all numeric features from the data set to standardize the predictors

```python
# Scale the numeric features
scaler = StandardScaler()
X_train_set_scaled = scaler.fit_transform(X_train)
X_test_set_scaled = scaler.fit_transform(X_test)
X_val_scaled = scaler.transform(X_val)
```

Now that the data has been sorted, we can create the MLP classifier and train the model. For each experiment, the number of hidden layers and number of neurons per hidden layer will be adjusted here.

```python
# Create an MLPClassifier without early stopping
mlp = MLPClassifier(hidden_layer_sizes=(50,50,50), max_iter=1000)

# Initialize lists to store training, validation, and test accuracies
train_accuracies = []
val_accuracies = []
test_accuracies = []

# Train the classifier and calculate accuracies at each iteration
# partial_fit() allows for incremental learning without retraining it from scratch
for i in range(1000):
    mlp.partial_fit(X_train_set_scaled, y_train, classes=np.unique(y_train))
    train_accuracies.append(mlp.score(X_train_set_scaled, y_train))
    val_accuracies.append(mlp.score(X_val_scaled, y_val))
    test_accuracies.append(mlp.score(X_test_set_scaled, y_test))
```

With the model trained, we can plot all of the data to view the results

```python
# Plot the training, validation, and test accuracies
plt.figure(figsize=(8, 6))
plt.plot(range(1, 1001), train_accuracies, label='Training accuracy')
plt.plot(range(1, 1001), val_accuracies, label='Validation accuracy')
plt.plot(range(1, 1001), test_accuracies, label='Test accuracy')
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

Plot the loss curve from the training

```python
plt.figure(figsize=(8, 6))
plt.plot(mlp.loss_curve_)
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.show()
```

Display the weights from the most recent training run

```python

for i, coef in enumerate(mlp.coefs_):
    print(f"Layer {i} - Weight matrix shape: {coef.shape}")

print()

for i, intercept in enumerate(mlp.intercepts_):
    print(f"Layer {i} - Intercept vector shape: {intercept.shape}")
```

Print out the overall training and test accuracy

```python
train_accuracy = mlp.score(X_train, y_train)
test_accuracy = mlp.score(X_test, y_test)

# Print the overall training and test accuracies
print(f"Overall Training Accuracy: {train_accuracy}")
print(f"Overall Test Accuracy: {test_accuracy}")
```

Display the confusion matrix

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Get the predicted labels for the test set
y_pred = mlp.predict(X_test)

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display the confusion matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', cbar=False)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```
